```python
# Used Packages
import nltk
nltk.download("wordnet")
from nltk.stem import WordNetLemmatizer
import re
```

# ***Preprocessing***

```python
# Reading files
with open("legit.txt", "r") as legit:
    legit = legit.read()
with open("plagiarized.txt", "r") as plagiarized:
    plagiarized = plagiarized.read()

# Remove prepositions, articles, etc
from gensim.parsing.preprocessing import remove_stopwords

filtered_legit = remove_stopwords(legit)
filtered_plagiarized = remove_stopwords(plagiarized)

# Turn to lowercase
filtered_legit = filtered_legit.lower()
filtered_plagiarized = filtered_plagiarized.lower()

# Split into sentences
legit_par = filtered_legit.split(". ")
plagiarized_par = filtered_plagiarized.split(". ")

# Define the pattern to match
pattern = r"[^\w\s]"


# Remove special characters
for i in range(len(legit_par)):
    legit_par[i] = re.sub(pattern, "", legit_par[i])

for i in range(len(plagiarized_par)):
    plagiarized_par[i] = re.sub(pattern, "", plagiarized_par[i])

# Spliting per sentence
for i in range(len(legit_par)):
    legit_par[i] = legit_par[i].split(" ")

for i in range(len(plagiarized_par)):
    plagiarized_par[i] = plagiarized_par[i].split(" ")

# Lemmatization
lemmatizer = WordNetLemmatizer()
for i in range(len(legit_par)):
    for j in range(len(legit_par[i])):
        legit_par[i][j] = lemmatizer.lemmatize(legit_par[i][j])

for i in range(len(plagiarized_par)):
    for j in range(len(plagiarized_par[i])):
        plagiarized_par[i][j] = lemmatizer.lemmatize(plagiarized_par[i][j])

# Join the sentences
for i in range(len(legit_par)):
    legit_par[i] = " ".join(legit_par[i])

for i in range(len(plagiarized_par)):
    plagiarized_par[i] = " ".join(plagiarized_par[i])
```
